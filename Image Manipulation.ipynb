```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img= cv.imread('noa.jpg')
img2= cv.imread('noa.jpg')
```

```python
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
scale_factor = 0.5
new_img = cv.resize(img, None, fx = scale_factor, fy = scale_factor, interpolation = cv.INTER_AREA)
plt.imshow(new_img)
plt.show()
```

```python
new_img.shape
```

```python
robo = new_img[ 0:250, 95:210]
```

```python
robo_red = cv.cvtColor(robo,cv.COLOR_BGR2RGB)
robo_gray = cv.cvtColor(robo_red,cv.COLOR_RGB2GRAY)
robo_gray3 = cv.cvtColor(robo_gray,cv.COLOR_GRAY2BGR)
img2[250:500, 35:150] = robo_gray3
img2[250:500, 475:590] = robo_red
plt.imshow(img2)
plt.show()
```
