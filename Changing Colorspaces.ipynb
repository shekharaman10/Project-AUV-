```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread('c1.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
img.size
```

```python
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('img',img)
```

```python
lower_blue = np.array([80, 50, 50])
upper_blue = np.array([130, 255, 255])

lower_red = np.array([160, 80, 20])
upper_red = np.array([179, 255, 255])
```

```python
# Threshold the HSV image to get only blue colors
mask = cv.inRange(hsv, lower_blue, upper_blue)
cv.imshow('Blue mask',mask)
```

```python
mask1= cv.inRange(hsv, lower_red, upper_red)
cv.imshow('Red mask',mask1)
```

```python
mask_combined = cv.bitwise_or(mask1, mask)
```

```python
result = cv.bitwise_and(img, img, mask=mask_combined)
```

```python
cv.imshow('Combined Mask', mask_combined)
cv.imshow('Result', result)
k = cv.waitKey(5) & 0xFF
if k == 27:  # 27 is the ASCII code for the 'Esc' key
    cv.destroyAllWindows()
```
